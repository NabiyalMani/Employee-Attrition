```python
data['Over18']=data['Over18'].replace({'Y':1,'N':0})
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
encoder=LabelEncoder()
for columns in encoding_cols:
    data[columns]=encoder.fit_transform(data[columns])
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus=RandomOverSampler(random_state=42)
x_over,y_over=rus.fit_resample(X,y)
      
print(Counter(y_over))
```

Training and Testing split


```python
# Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_over, y_over, test_size=0.2, random_state=42)
```

Logistic Regrassion model


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg= LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
parameters={'penalty':['l1','l2'],
            'C' : np.logspace(-3,3,7),
            'solver' : ['newton-cg','lbfgs','liblinear']}
```

```python
logreg= LogisticRegression()
clf=GridSearchCV(logreg,param_grid=parameters, scoring='accuracy', cv=10)
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
prediction=clf.predict(X_test)
```

```python
cnf_matrix=confusion_matrix(y_test,prediction)
```

```python
print("Accuracy Score-",accuracy_score(y_test, prediction))
```
